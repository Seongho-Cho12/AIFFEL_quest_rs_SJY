```python
#필요한 패키지 import 하기
import os
import cv2  #OpenCV 라이브러리, cv 관련 프로그래밍을 쉽게 할 수 있도록 도와주는 라이브러리
import matplotlib.pyplot as plt
import numpy as np
import dlib #이미지 처리 및 기계 학습, 얼굴인식 등을 할 수 있는 c++로 개발된 고성능 라이브러리
```

## 1. 얼굴 검출 & 랜드마크 검출하기

```python
#경로 확인
import os
print(os.getcwd())
```

```python
#이미지 읽기
my_image_path = '/workspace/userdisk/AIFFEL_quest_cr/AIFFEL_quest_rs/Exploration/Ex03/face.png'
img_bgr = cv2.imread(my_image_path) #opencv로 이미지 불러오기
plt.imshow(img_bgr)
```

```python
#RGB 이미지로 변환
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
#dlib의 face detector를 사용하여 얼굴 찾기


#detector 선언
detector_hog = dlib.get_frontal_face_detector()

#detector_hog를 이용하여 얼굴의 bounding box 추출
dlib_rects = detector_hog(img_rgb, 1)
```

```python
#찾은 얼굴 영역 박스 리스트 확인하기
print(dlib_rects)
for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()
    
    cv2.rectangle(img_rgb, (l,t), (r,b), (0, 255, 0), 2, lineType = cv2.LINE_AA)
    
plt.imshow(img_rgb)
plt.show()
```

```python
#저장한 랜드마크 모델 불러오기
model_path = '/workspace/userdisk/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
    # 저장한 landmark 모델의 주소를 model_path 변수에 저장
landmark_predictor = dlib.shape_predictor(model_path)
    # dlib 라이브러리의 shape_predictor 함수를 이용하여 모델을 불러옴
    # landmark_predictor는 RGB이미지와 dlib.rectangle를 입력 받고 dlib.full_object_detection를 반환
    # dlib.rectangle: 내부를 예측하는 박스
    # dlib.full_object_detection: 각 구성 요소의 위치와, 이미지 상의 객체의 위치를 나타냄
```

```python
#landmark_predictor로 rgb 이미지와 dlib.rectangle을 입력받아 
#dlib.full_object_detection 반환

list_landmarks = []

for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)
    
print(len(list_landmarks[0]))


#랜드마크 출력
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_rgb, point, 6, (0, 255, 255), -1)
        
plt.imshow(img_rgb)
plt.show()
```

## 2. 스티커 적용 위치 확인하기
- 고양이 수염이 적용될 위치를 landmark를 사용하여 계산
- 코끝 33번 보다 좋은 위치는 무엇일까?

```python
#list_landmarks의 68개의 랜드마크 값이 저장되어있다. 
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print(landmark[33]) #
    x = landmark[33][0]
    y = landmark[33][1]
    w = h = dlib_rect.width() #얼굴 가로영역을 차지하는 픽셀의 수
    print (f'(x,y) : ({x},{y})')
    print (f'(w,h) : ({w},{h})')
```

## 3. 스티커 적용하기

```python
sticker_path = '/workspace/userdisk/AIFFEL_quest_cr/AIFFEL_quest_rs/Exploration/Ex03/cat-whiskers.png' # 고양이 수염 이미지 경로
img_sticker = cv2.imread(sticker_path) # 스티커 이미지를 불러옵니다 // cv2.imread(이미지 경로) → image객체 행렬을 반환
img_sticker = cv2.resize(img_sticker, (w,h)) # 스티커 이미지 조정 → w,h는 얼굴 영역의 가로를 차지하는 픽셀의 수(187) // cv2.resize(image객체 행렬, (가로 길이, 세로 길이))
print (img_sticker.shape) #
```

```python
refined_x = x - w // 2 -1
refined_y = y - w // 2 -5
# 즉, refined_x, refined_y값에서 이미지가 시작됨
print (f'(x,y) : ({refined_x},{refined_y})')
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
sticker_area = img_rgb[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_rgb[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
print("슝~")
```
