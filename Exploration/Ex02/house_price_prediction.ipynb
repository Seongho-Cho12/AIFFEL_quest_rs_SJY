## 3. 모델 학습

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor


random_state = 123
gboost = GradientBoostingRegressor(random_state=random_state)
xgboost = XGBRegressor(random_state=random_state)
lightgbm = LGBMRegressor(random_state=random_state)
rdforest = RandomForestRegressor(random_state=random_state)

models = [gboost, xgboost, lightgbm, rdforest]
```

```python
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
#모델 별 성능 비교
df = {}
def get_scores(models, train, y):
    for model in models:
        # 모델 이름 획득
        model_name = model.__class__.__name__

        # train, test 데이터셋 분리
        # random_state를 사용하여 고정하고 train과 test 셋의 비율은 8:2로 합니다.
        x_train, x_test, y_train, y_test = train_test_split(train, y, test_size = 0.2, random_state = 123)

        # 모델 학습
        model.fit(x_train, y_train)

        # 예측
        y_pred = model.predict(x_test)

        # 예측 결과의 rmse값 저장
        df[model_name] = rmse(y_test, y_pred)

         # data frame에 저장
        score_df = pd.DataFrame(df, index=['RMSE']).T.sort_values('RMSE', ascending=False)

    return score_df
get_scores(models, train, y)
```

```python
# model, train, y, param_grid={'n_estimators': [50, 100], 'max_depth': [1, 10], #'learning_rate' :[0.3, 0.5], 'booster' :['dart']}, verbose=2, n_jobs=-1):

def get_gs_result(model, train, y, param_grid, verbose=2, n_jobs=-1):

    grid_model = GridSearchCV(
        model, param_grid=param_grid,
        scoring='neg_mean_squared_error',
        cv=5, verbose=verbose, n_jobs=n_jobs)

    grid_model.fit(train, y)

    params = grid_model.cv_results_['params']
    score = grid_model.cv_results_['mean_test_score']

    results = pd.DataFrame(params)
    results['score'] = score

    results['RMSLE'] = np.sqrt(-1 * results['score'])
    results = results.sort_values('RMSLE')

    return results
```

```python
#LGB<Regressor의  하이처 파라미터 튜닝
param_grid={'n_estimators': [50, 100], 'max_depth': [10, 100], 'learning_rate' :[0.3, 0.5], 'booster' :['dart', 'gbdt']}
get_gs_result(
    LGBMRegressor(random_state=random_state), train, y, param_grid)
```

```python
#XGB 하이퍼파라미터 튜닝
param_grid={'n_estimators': [100, 200], 'max_depth': [50, 100], 'learning_rate' :[0.1, 0.3], 'booster' :['dart']}

get_gs_result(
    XGBRegressor(random_state=random_state), train, y, param_grid)
```

```python
#rf, 하이퍼 파라미터 튜닝
param_grid={'n_estimators': [50, 100], 'max_depth': [1, 10]}

get_gs_result(
    RandomForestRegressor(random_state=random_state), train, y, param_grid)
```

```python
#Gradientboosting 하이퍼 파라미터 튜닝
param_grid={'n_estimators': [50, 100], 'max_depth': [1, 10]}
get_gs_result(
    GradientBoostingRegressor(random_state=random_state), train, y, param_grid)
```

```python
#ensemble
from sklearn.ensemble import StackingRegressor
X_train, X_test, y_train, y_test = train_test_split(train, y, random_state=123, test_size=0.2)

#gboost = GradientBoostingRegressor(max_depth = 10, n_estimators = 100,random_state=123)
xgboost = XGBRegressor(max_depth = 10, n_estimators = 50, booster = 'dart', learning_rate = 0.3,random_state=123)
lightgbm = LGBMRegressor(max_depth =10 , n_estimators=100, boosting_type = 'dart', learning_rate = 0.3 ,random_state=123)
#rdforest = RandomForestRegressor(max_depth = 10, n_estimators = 100,random_state=123)


print(type(gboost))
print(type(xgboost))
print(type(lightgbm))
print(type(rdforest))

stack_models = [
    #('gb', gboost),
    ('xgb', xgboost),
    ('lgbm', lightgbm),
    #('rf', rdforest)
]

stack_model = StackingRegressor(estimators=stack_models, final_estimator=lightgbm)
stack_model.fit(X_train, y_train)
```

```python
#stacking 모델의 성능
y_pred = stack_model.predict(X_test)
rmse(y_pred, y_test)
```
